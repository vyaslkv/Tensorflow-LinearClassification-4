```python
import pandas as pd
```

```python
df = pd.read_csv('census_data.csv')
```

```python
df.head()
```

```python
df['income_bracket'].unique()
```

```python
def label_fix(label):
    if label == ' <=50K':
        return 0
    else:
        return 1
```

```python
df['income_bracket'] = df['income_bracket'].apply(label_fix)
```

```python
from sklearn.model_selection import train_test_split
```

```python
df.columns
```

```python
x_data = df.drop('income_bracket',axis=1)
y = df['income_bracket']
```

```python
X_train, X_eval, y_train, y_eval = train_test_split(x_data, y, test_size=0.3, random_state=101)
```

```python
df.columns
```

```python
df.head()
```

```python
import tensorflow as tf
```

```python
gender = tf.feature_column.categorical_column_with_vocabulary_list("gender", ["Female", "Male"])
occupation = tf.feature_column.categorical_column_with_hash_bucket("occupation", hash_bucket_size=1000)
marital_status = tf.feature_column.categorical_column_with_hash_bucket("marital_status", hash_bucket_size=1000)
relationship = tf.feature_column.categorical_column_with_hash_bucket("relationship", hash_bucket_size=1000)
education = tf.feature_column.categorical_column_with_hash_bucket("education", hash_bucket_size=1000)
workclass = tf.feature_column.categorical_column_with_hash_bucket("workclass", hash_bucket_size=1000)
native_country = tf.feature_column.categorical_column_with_hash_bucket("native_country", hash_bucket_size=1000)
age = tf.feature_column.numeric_column("age")
education_num = tf.feature_column.numeric_column("education_num")
capital_gain = tf.feature_column.numeric_column("capital_gain")
capital_loss = tf.feature_column.numeric_column("capital_loss")
hours_per_week = tf.feature_column.numeric_column("hours_per_week")
```

```python
feat_cols = [gender,occupation,marital_status,relationship,education,workclass,native_country,
            age,education_num,capital_gain,capital_loss,hours_per_week]
```

```python
input_func = tf.estimator.inputs.pandas_input_fn(x=X_train, y=y_train, batch_size=10, num_epochs=None, shuffle=True)
```

```python
model = tf.estimator.LinearClassifier(feature_columns = feat_cols)
```

```python
model.train(input_fn=input_func, steps=5000)
```

```python
eval_input_func = tf.estimator.inputs.pandas_input_fn(X_eval, y_eval, batch_size=10, num_epochs=1, shuffle=False)
```

```python
model.evaluate(input_fn = eval_input_func, steps=5000)
```

```python
pred_input_func = tf.estimator.inputs.pandas_input_fn(X_eval, batch_size=10, num_epochs=1, shuffle=False)
```

```python
model.predict(input_fn = pred_input_func)
```

```python
predictions = list(model.predict(input_fn = pred_input_func))
```

```python
predictions
```

```python
pred_list = []
for pred in predictions:
    pred_list.append(pred['class_ids'][0])
```

```python
pred_list
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_eval, pred_list))
```
